**Hint:** Assume the first input to be the largest value initially and the largest value’s location to be 0.

**Sample Input:**\
 7, 13, 2, 10, 6

**Sample Output:**\

===================================================================

```python
#to do
user_input = input("Please enter the number: ")
the_list = user_input.split(",")

for index in range (5) :
  the_list[index] = int(the_list[index])

the_list = [int(index) for index in the_list]
large = the_list[0]
index = 0

for counter in range (len(the_list)) :
  if (large <= the_list[counter]) :
    large = the_list[counter]
    index = counter

print("My list:", the_list)
print("Largest number in the list is", large, "which was found at index", index,".")
```
