Doubled number: a number where every digit is double of its immediate left digit is called a Doubled number. (e.g. 1248 is a Doubled number).

Now write a python program that takes a number from the user and checks whether it is a Doubled number or not. 

========================================================
Sample Input1
1248
Sample Output1
1248 YES


Explanation1:
2 is double of 1, 4 is double of 2, 8 is double of 4.

================================
================================
Sample Input2
11
Sample Output2
11 NO
Explanation2:
Digits are not doubled
================================
Sample Input3:
48
Sample Output3:
48 YES

Explanation3:
8 is double of 4

```python
variable_1="1248"

for digits in range (len(variable_1)) :
  print(variable_1[digits])
```

```python
number="1248"


for digits in range (len(number)) :
  num=digits
  num2=digits*2
  if num2%2 != num :
    print(number, "Yes")
    break
  
else :
    print(number, "No")
```
