```python
# Task 2

word = input("Please enter a word: ")
new = ''
length = len(word)

for index in range (len(word)) :
  if len(word) % 2 == 0 :
    if index == length // 2 or index == (length//2)-1 :
      new = new + str(ord(word[index])) + ' '
    else :
      new = new + word[index] + ' '
  elif len(word) % 2 == 1 :
    if index == ((len(word)+1)//2)-1 or index == (len(word)+1)//2 or index == ((len(word)+1)//2) + 1 :
      new = new + str(ord(word[index])) + ' '
    else :
      new = new + word[index] + ' '

print(new)
```
